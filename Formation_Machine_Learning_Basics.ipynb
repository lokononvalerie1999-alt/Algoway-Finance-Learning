```python
# Calcule le TOTAL de tout ton portefeuille

actions = ["Orange", "Safarel", "Ecobank"]
prix_achat = [1500, 2000, 5000]
quantites = [10, 5, 2]
prix_actuels = [1650, 1950, 4800]

total_investi = 0
total_actuel = 0

for i in range(len(actions)):
    invested = prix_achat[i] * quantites[i]
    current = prix_actuels[i] * quantites[i]
    
    total_investi += invested  # Accumule
    total_actuel += current    # Accumule

gain_total = total_actuel - total_investi
rendement_total = (gain_total / total_investi) * 100

print(f"Total investi: {total_investi} FCFA")
print(f"Total actuel: {total_actuel} FCFA")
print(f"Gain/Perte: {gain_total} FCFA ({rendement_total:.2f}%)")
```

```python
# ========== JOUR 3: BOUCLES WHILE ==========

# Scénario: Un prix monte chaque jour jusqu'à atteindre 2000

prix = 1500
jour = 0
prix_objectif = 2000

print("Simulation: Prix qui monte")
print("─" * 40)

while prix < prix_objectif:
    jour += 1
    prix += 100  # Le prix monte de 100 chaque jour
    print(f"Jour {jour}: Prix = {prix} FCFA")

print(f"✅ Objectif atteint le jour {jour}!")
```

```python
# Scénario: Tu achètes des actions tant que le prix est bon

prix_courant = 1800
argent_disponible = 20000
prix_achat = 0
nombre_actions = 0

print("Achat d'actions tant que possible")
print("─" * 40)

while argent_disponible >= prix_courant:
    nombre_actions += 1
    argent_disponible -= prix_courant
    prix_achat += prix_courant
    
    print(f"Action {nombre_actions}: Acheté à {prix_courant} FCFA")
    print(f"  Argent restant: {argent_disponible} FCFA")

print(f"\n✅ Achat terminé!")
print(f"Nombre d'actions: {nombre_actions}")
print(f"Argent utilisé: {prix_achat} FCFA")
print(f"Argent restant: {argent_disponible} FCFA")
```

```python
# ========== SIMULATION: PERTE PROGRESSIVE ==========

prix_initial = 1500
prix_courant = prix_initial
limite_perte = 1300  # Si prix < 1300, tu vends (STOP-LOSS)
jour = 0

# BOUCLE: Tant que le prix est SUPÉRIEUR à la limite
while prix_courant > limite_perte:
    jour += 1                    # Ajoute 1 jour
    prix_courant -= 50           # Le prix baisse de 50
    print(f"Jour {jour}: Prix = {prix_courant} FCFA")

# Une fois sorti de la boucle, affiche le résumé
print("\n" + "="*50)
print(f"⚠️  ALERTE STOP-LOSS!")
print(f"Le prix a atteint {prix_courant} FCFA le jour {jour}")
print(f"Perte totale: {prix_initial - prix_courant} FCFA")
print("="*50)
```

```python
# ========== TÂCHE 3: COMPTER LES ACTIONS EN GAIN ==========

actions = ["Orange", "Safarel", "Ecobank", "SITAB"]
prix_achat = [1500, 2000, 5000, 300]
prix_actuels = [1650, 1950, 4800, 350]

print("="*60)
print("TÂCHE 3: Compter les actions en GAIN")
print("="*60)

# Crée des variables pour compter
en_gain = 0
en_perte = 0

# Boucle sur chaque action
for i in range(len(actions)):
    action = actions[i]
    prix_achat_val = prix_achat[i]
    prix_actuel_val = prix_actuels[i]
    
    # Calcule le gain/perte
    gain = prix_actuel_val - prix_achat_val
    
    # Affiche chaque action
    if gain > 0:
        en_gain += 1
        print(f"✅ {action}: +{gain} FCFA (EN GAIN)")
    else:
        en_perte += 1
        print(f"❌ {action}: {gain} FCFA (EN PERTE)")

# RÉSUMÉ
print("\n" + "="*60)
print("📊 RÉSUMÉ:")
print(f"  Actions en GAIN: {en_gain}")
print(f"  Actions en PERTE: {en_perte}")
print(f"  Total: {en_gain + en_perte}")
print("="*60)
```

```python
# ========== JOUR 5: ANALYSER 10 ACTIONS ==========
# Combine: Listes + Boucles + Conditions + Calculs

import random  # Pour générer des données aléatoires

print("="*80)
print("ANALYSEUR DE PORTEFEUILLE - 10 ACTIONS")
print("="*80)

# ===== TES 10 ACTIONS =====

actions = [
    "Orange",
    "Safarel",
    "Ecobank",
    "SITAB",
    "Benin Telecom",
    "Total Benin",
    "ONATEL",
    "COGEBANQUE",
    "CITBANK",
    "NSIA Banque"
]

# Prix d'achat (données réalistes ou fictives)
prix_achat = [1500, 2000, 5000, 300, 1200, 3500, 800, 4000, 2500, 1800]

# Quantités
quantites = [10, 5, 2, 20, 15, 3, 25, 4, 8, 12]

# Prix actuels (change ces valeurs pour simuler différents scénarios)
prix_actuels = [1650, 1950, 4800, 350, 1300, 3600, 750, 4200, 2400, 1900]

# ===== ANALYSE =====

print(f"\n{'Rang':<5} {'Action':<20} {'Inv. (FCFA)':<15} {'Actuel (FCFA)':<15} {'Gain/Perte':<15} {'%':<8} {'Status':<10}")
print("─"*98)

total_investi = 0
total_actuel = 0
en_gain = 0
en_perte = 0

for i in range(len(actions)):
    action = actions[i]
    prix = prix_achat[i]
    quantite = quantites[i]
    prix_actuel = prix_actuels[i]
    
    # Calculs
    investi = prix * quantite
    actuel = prix_actuel * quantite
    gain = actuel - investi
    rendement = (gain / investi) * 100
    
    # Status
    if gain > 0:
        status = "✅ GAIN"
        en_gain += 1
    else:
        status = "❌ PERTE"
        en_perte += 1
    
    # Accumule
    total_investi += investi
    total_actuel += actuel
    
    # Affiche
    print(f"{i+1:<5} {action:<20} {investi:<15.0f} {actuel:<15.0f} {gain:<15.0f} {rendement:<8.2f} {status:<10}")

# ===== RÉSUMÉ =====

gain_total = total_actuel - total_investi
rendement_total = (gain_total / total_investi) * 100

print("─"*98)
print(f"{'TOTAL':<5} {'':<20} {total_investi:<15.0f} {total_actuel:<15.0f} {gain_total:<15.0f} {rendement_total:<8.2f}")
print("="*98)

# ===== STATISTIQUES =====

print(f"\n📊 STATISTIQUES:")
print(f"─" * 50)
print(f"Nombre d'actions: {len(actions)}")
print(f"Actions en GAIN: {en_gain}")
print(f"Actions en PERTE: {en_perte}")
print(f"Total investi: {total_investi:,.0f} FCFA")
print(f"Valeur actuelle: {total_actuel:,.0f} FCFA")
print(f"Gain/Perte total: {gain_total:,.0f} FCFA")
print(f"Rendement global: {rendement_total:+.2f}%")

# ===== RECOMMANDATIONS =====

print(f"\n💡 RECOMMANDATIONS:")
print(f"─" * 50)

if rendement_total > 5:
    print("✅ Ton portefeuille se porte TRÈS BIEN! Continue!")
elif rendement_total > 0:
    print("📈 Ton portefeuille est en GAIN. Maintiens le cap!")
elif rendement_total > -5:
    print("⚠️ Ton portefeuille est en légère PERTE. À surveiller.")
else:
    print("❌ Ton portefeuille est en PERTE importante. Réévalue ta stratégie!")

print("="*98)
```

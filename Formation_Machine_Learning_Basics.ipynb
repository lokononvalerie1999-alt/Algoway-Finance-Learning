```python
# Calcule le TOTAL de tout ton portefeuille

actions = ["Orange", "Safarel", "Ecobank"]
prix_achat = [1500, 2000, 5000]
quantites = [10, 5, 2]
prix_actuels = [1650, 1950, 4800]

total_investi = 0
total_actuel = 0

for i in range(len(actions)):
    invested = prix_achat[i] * quantites[i]
    current = prix_actuels[i] * quantites[i]
    
    total_investi += invested  # Accumule
    total_actuel += current    # Accumule

gain_total = total_actuel - total_investi
rendement_total = (gain_total / total_investi) * 100

print(f"Total investi: {total_investi} FCFA")
print(f"Total actuel: {total_actuel} FCFA")
print(f"Gain/Perte: {gain_total} FCFA ({rendement_total:.2f}%)")
```

```python
# ========== JOUR 3: BOUCLES WHILE ==========

# Scénario: Un prix monte chaque jour jusqu'à atteindre 2000

prix = 1500
jour = 0
prix_objectif = 2000

print("Simulation: Prix qui monte")
print("─" * 40)

while prix < prix_objectif:
    jour += 1
    prix += 100  # Le prix monte de 100 chaque jour
    print(f"Jour {jour}: Prix = {prix} FCFA")

print(f"✅ Objectif atteint le jour {jour}!")
```

```python
# Scénario: Tu achètes des actions tant que le prix est bon

prix_courant = 1800
argent_disponible = 20000
prix_achat = 0
nombre_actions = 0

print("Achat d'actions tant que possible")
print("─" * 40)

while argent_disponible >= prix_courant:
    nombre_actions += 1
    argent_disponible -= prix_courant
    prix_achat += prix_courant
    
    print(f"Action {nombre_actions}: Acheté à {prix_courant} FCFA")
    print(f"  Argent restant: {argent_disponible} FCFA")

print(f"\n✅ Achat terminé!")
print(f"Nombre d'actions: {nombre_actions}")
print(f"Argent utilisé: {prix_achat} FCFA")
print(f"Argent restant: {argent_disponible} FCFA")
```

```python
# ========== SIMULATION: PERTE PROGRESSIVE ==========

prix_initial = 1500
prix_courant = prix_initial
limite_perte = 1300  # Si prix < 1300, tu vends (STOP-LOSS)
jour = 0

# BOUCLE: Tant que le prix est SUPÉRIEUR à la limite
while prix_courant > limite_perte:
    jour += 1                    # Ajoute 1 jour
    prix_courant -= 50           # Le prix baisse de 50
    print(f"Jour {jour}: Prix = {prix_courant} FCFA")

# Une fois sorti de la boucle, affiche le résumé
print("\n" + "="*50)
print(f"⚠️  ALERTE STOP-LOSS!")
print(f"Le prix a atteint {prix_courant} FCFA le jour {jour}")
print(f"Perte totale: {prix_initial - prix_courant} FCFA")
print("="*50)
```

```python
# ========== JOUR 2: AJOUTER DES COLONNES ==========

import pandas as pd

# Crée le DataFrame de base
donnees = {
    'Action': ['Orange', 'Safarel', 'Ecobank'],
    'Prix_Achat': [1500, 2000, 5000],
    'Quantite': [10, 5, 2],
    'Prix_Actuel': [1650, 1950, 4800]
}
df = pd.DataFrame(donnees)

print("="*80)
print("JOUR 2: AJOUTER DES COLONNES CALCULÉES")
print("="*80)

print("\n📊 DATAFRAME ORIGINAL:")
print(df)

# ===== COLONNE 1: Total Investi =====
print("\n" + "─"*80)
print("1️⃣  AJOUTER: Total_Investi = Prix_Achat × Quantite")
print("─"*80)

df['Total_Investi'] = df['Prix_Achat'] * df['Quantite']
print(df)

# ===== COLONNE 2: Valeur Actuelle =====
print("\n" + "─"*80)
print("2️⃣  AJOUTER: Valeur_Actuelle = Prix_Actuel × Quantite")
print("─"*80)

df['Valeur_Actuelle'] = df['Prix_Actuel'] * df['Quantite']
print(df)

# ===== COLONNE 3: Gain/Perte =====
print("\n" + "─"*80)
print("3️⃣  AJOUTER: Gain_Perte = Valeur_Actuelle - Total_Investi")
print("─"*80)

df['Gain_Perte'] = df['Valeur_Actuelle'] - df['Total_Investi']
print(df)

# ===== COLONNE 4: Rendement % =====
print("\n" + "─"*80)
print("4️⃣  AJOUTER: Rendement_% = (Gain_Perte / Total_Investi) × 100")
print("─"*80)

df['Rendement_%'] = (df['Gain_Perte'] / df['Total_Investi']) * 100
print(df)

# ===== RÉSULTAT FINAL =====
print("\n" + "="*80)
print("📊 TABLEAU FINAL COMPLET:")
print("="*80)
print(df)

# ===== AFFICHAGE PLUS LISIBLE =====
print("\n" + "="*80)
print("📋 AFFICHAGE FORMATÉ:")
print("="*80)

for i in range(len(df)):
    action = df.loc[i, 'Action']
    investi = df.loc[i, 'Total_Investi']
    gain = df.loc[i, 'Gain_Perte']
    rendement = df.loc[i, 'Rendement_%']
    
    print(f"\n{action}:")
    print(f"  Total investi: {investi:>10,.0f} FCFA")
    print(f"  Gain/Perte:   {gain:>10,.0f} FCFA")
    print(f"  Rendement:    {rendement:>10.2f}%")
```

```python
# ========== JOUR 3: MOYENNES MOBILES - FINAL CORRIGÉ ==========

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("JOUR 3: MOYENNES MOBILES - NETTOYAGE FINAL")
print("="*80)

# ===== FONCTION DE TÉLÉCHARGEMENT CORRECTE =====

def telecharger_et_nettoyer(ticker, nom=""):
    """
    Télécharge UNE action et NETTOIE CORRECTEMENT le MultiIndex
    """
    
    if not nom:
        nom = ticker
    
    print(f"  📊 {nom}...", end=" ", flush=True)
    
    try:
        # Télécharge
        data = yf.download(ticker, start='2024-01-01', end='2026-04-29', progress=False)
        
        if len(data) == 0:
            print("❌ (pas de données)")
            return None
        
        # ✅ NETTOYAGE CORRECT DU MULTIINDEX
        print(f"(structure", end=" ", flush=True)
        
        if isinstance(data.columns, pd.MultiIndex):
            # Structure de yfinance:
            # Level 0: Type de donnée (Close, High, Low, etc.)
            # Level 1: Ticker (AAPL, MSFT, etc.)
            
            # On veut GARDER le niveau 0 (les types de données)
            # et ENLEVER le niveau 1 (le ticker)
            
            # Solution: flatten et garder le premier niveau
            data.columns = data.columns.get_level_values(0)
        
        print("ok)", end=" ", flush=True)
        print(f"✅ ({len(data)} jours)")
        return data
        
    except Exception as e:
        print(f"❌ ({str(e)[:40]})")
        return None

# ===== TÉLÉCHARGER =====

print("\n" + "─"*80)
print("1️⃣  TÉLÉCHARGER ET NETTOYER")
print("─"*80)

print("\n🔍 Recherche des données...")

tickers_test = [
    ('ORNT.PA', 'Orange'),
    ('AAPL', 'Apple'),
    ('MSFT', 'Microsoft'),
    ('GOOGL', 'Google'),
]

data = None
premier = None

for ticker, nom in tickers_test:
    data = telecharger_et_nettoyer(ticker, nom)
    if data is not None:
        premier = nom
        break

if data is None:
    print("\n❌ Impossible de télécharger!")
    exit()

print(f"\n✅ Utilisation de: {premier}")

# ===== VÉRIFIER LA STRUCTURE =====

print("\n" + "─"*80)
print("2️⃣  VÉRIFICATION DE LA STRUCTURE")
print("─"*80)

print(f"\n  Colonnes: {data.columns.tolist()}")
print(f"  Nombre de colonnes: {len(data.columns)}")
print(f"  Nombre de lignes: {len(data)}")
print(f"  Est MultiIndex? {isinstance(data.columns, pd.MultiIndex)}")

# Vérification critique
if 'Close' not in data.columns:
    print(f"\n  ❌ ERREUR: Colonne 'Close' introuvable!")
    print(f"  Colonnes disponibles: {data.columns.tolist()}")
    print(f"\n  💡 Solution: Les colonnes ne sont pas au bon format")
    print(f"  Essayons une méthode alternative...")
    
    # Méthode alternative: recréer les colonnes correctement
    if len(data.columns) > 0:
        # Prends les noms du MultiIndex d'avant
        print(f"\n  Reconstruction en cours...")
        # Réinitialise les colonnes
        data = yf.download(
            'AAPL',
            start='2024-01-01',
            end='2026-04-29',
            progress=False,
            auto_adjust=False  # ← Important!
        )
        
        if isinstance(data.columns, pd.MultiIndex):
            # Garde seulement le premier niveau
            data.columns = data.columns.get_level_values(0)
        
        print(f"  ✅ Colonnes maintenant: {data.columns.tolist()}")
else:
    print(f"  ✅ Colonne 'Close' trouvée!")

# ===== AFFICHER LES PREMIÈRES LIGNES =====

print("\n" + "─"*80)
print("3️⃣  VÉRIFICATION DES DONNÉES")
print("─"*80)

print(f"\n📊 Premières 3 lignes:")
print(data[['Open', 'High', 'Low', 'Close', 'Volume']].head(3).to_string())

# ===== CALCULER LES MOYENNES MOBILES =====

print("\n" + "─"*80)
print("4️⃣  CALCULER LES MOYENNES MOBILES")
print("─"*80)

print(f"\n  Calcul en cours...")

# Vérification finale
if 'Close' not in data.columns:
    print(f"  ❌ ERREUR CRITIQUE: 'Close' introuvable!")
    print(f"  Colonnes: {data.columns.tolist()}")
    exit()

# Calcul
data['MA20'] = data['Close'].rolling(window=20).mean()
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA100'] = data['Close'].rolling(window=100).mean()

print(f"  ✅ MA20 calculée")
print(f"  ✅ MA50 calculée")
print(f"  ✅ MA100 calculée")

# ===== RÉSULTATS =====

print("\n" + "─"*80)
print("5️⃣  RÉSULTATS")
print("─"*80)

print(f"\n📊 Dernières 5 lignes:")

affichage = data[['Close', 'MA20', 'MA50', 'MA100']].tail(5)
print(affichage.to_string())

# ===== INTERPRÉTATION =====

print("\n" + "─"*80)
print("6️⃣  INTERPRÉTATION")
print("─"*80)

# Accès correct aux données
prix_actuel = float(data['Close'].iloc[-1])
ma20_actuel = float(data['MA20'].iloc[-1])
ma50_actuel = float(data['MA50'].iloc[-1])

print(f"\n💰 Situation Actuelle:")
print(f"  Prix actuel:      {prix_actuel:.2f}")
print(f"  Moyenne 20 jours: {ma20_actuel:.2f}")
print(f"  Moyenne 50 jours: {ma50_actuel:.2f}")

print(f"\n📈 Analyse:")

if prix_actuel > ma20_actuel:
    print(f"  ✅ Prix DESSUS MA20 → Tendance HAUSSIÈRE court terme")
else:
    print(f"  ❌ Prix SOUS MA20 → Tendance BAISSIÈRE court terme")

if prix_actuel > ma50_actuel:
    print(f"  ✅ Prix DESSUS MA50 → Tendance HAUSSIÈRE moyen terme")
else:
    print(f"  ❌ Prix SOUS MA50 → Tendance BAISSIÈRE moyen terme")

if ma20_actuel > ma50_actuel:
    print(f"\n  🔔 SIGNAL: MA20 > MA50 → Tendance HAUSSIÈRE générale ✅")
else:
    print(f"\n  🔔 SIGNAL: MA20 < MA50 → Tendance BAISSIÈRE générale ❌")

# ===== CROISEMENTS =====

print("\n" + "─"*80)
print("7️⃣  CROISEMENTS (GOLDEN CROSS / DEATH CROSS)")
print("─"*80)

# Détecte les croisements
ma20_series = data['MA20'].dropna()
ma50_series = data['MA50'].dropna()

# Index communs
common_index = ma20_series.index.intersection(ma50_series.index)
ma20_clean = ma20_series.loc[common_index]
ma50_clean = ma50_series.loc[common_index]

# Crée une colonne pour la différence
croisement_data = pd.DataFrame({
    'MA20': ma20_clean,
    'MA50': ma50_clean
})
croisement_data['MA_Diff'] = croisement_data['MA20'] - croisement_data['MA50']
croisement_data['Croisement'] = (
    (croisement_data['MA_Diff'].shift(1) * croisement_data['MA_Diff'] < 0)
).astype(int)

croisements = croisement_data[croisement_data['Croisement'] == 1]

print(f"\n🔔 CROISEMENTS DÉTECTÉS: {len(croisements)}")

if len(croisements) > 0:
    print("\nDerniers croisements:")
    for idx, row in croisements.tail(5).iterrows():
        if row['MA_Diff'] > 0:
            signal = "📈 GOLDEN CROSS (Achat)"
        else:
            signal = "📉 DEATH CROSS (Vente)"
        print(f"  {idx.date()}: {signal}")
else:
    print("  Aucun croisement détecté")

# ===== GRAPHIQUE =====

print("\n" + "─"*80)
print("8️⃣  GRAPHIQUE")
print("─"*80)

plt.figure(figsize=(14, 7))

plt.plot(data.index, data['Close'], linewidth=2, label='Prix', color='black', alpha=0.7)
plt.plot(data.index, data['MA20'], linewidth=2, label='MA20', color='orange')
plt.plot(data.index, data['MA50'], linewidth=2, label='MA50', color='red')
plt.plot(data.index, data['MA100'], linewidth=2, label='MA100', color='purple', alpha=0.7)

plt.title(f'{premier} - Prix + Moyennes Mobiles', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Prix')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Graphique affiché!")

# ===== RÉSUMÉ FINAL =====

print("\n" + "="*80)
print("✅ JOUR 3 TERMINÉ!")
print("="*80)

print(f"""
📊 Résumé:
  • Données: {premier}
  • Jours: {len(data)}
  • Colonnes: {data.columns.tolist()[:3]}...
  • MA20, MA50, MA100: ✅ Calculées
  • Structure: ✅ Correcte

🎯 Prochaine étape:
  Jour 4: Signaux de Trading
""")

print("="*80)
```

```python
# ========== JOUR 4: SIGNAUX DE TRADING - VERSION FINALE ==========

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("JOUR 4: SIGNAUX DE TRADING")
print("="*80)

# ===== CONCEPT =====

print("""
📊 SIGNAUX DE TRADING:

Un signal = Indication pour ACHETER ou VENDRE

SIGNAL BUY (Achat) 🟢:
  ✅ Prix croise MA20 vers le haut
  ✅ MA20 > MA50 (tendance haussière)
  ✅ RSI < 70 (pas surachat)

SIGNAL SELL (Vente) 🔴:
  ❌ Prix croise MA20 vers le bas
  ❌ MA20 < MA50 (tendance baissière)
  ❌ RSI > 30 (pas survendu)

RSI (Relative Strength Index):
  • 0-30:   Survendu (opportunité d'achat)
  • 30-70:  Zone normale
  • 70-100: Surachat (attention!)
""")

# ===== FONCTION DE TÉLÉCHARGEMENT =====

def telecharger_et_nettoyer(ticker, nom=""):
    """Télécharge une action et nettoie le MultiIndex"""
    
    if not nom:
        nom = ticker
    
    print(f"  📊 {nom}...", end=" ", flush=True)
    
    try:
        data = yf.download(ticker, start='2024-01-01', end='2026-04-29', progress=False)
        
        if len(data) == 0:
            print("❌")
            return None
        
        # ✅ NETTOYAGE
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = data.columns.get_level_values(0)
        
        print(f"✅ ({len(data)} jours)")
        return data
        
    except Exception as e:
        print(f"❌")
        return None

# ===== TÉLÉCHARGER =====

print("\n" + "─"*80)
print("1️⃣  TÉLÉCHARGER LES DONNÉES")
print("─"*80)

print("\n🔍 Recherche des données...")

tickers_test = [
    ('AAPL', 'Apple'),
    ('MSFT', 'Microsoft'),
    ('GOOGL', 'Google'),
]

data = None
premier = None

for ticker, nom in tickers_test:
    data = telecharger_et_nettoyer(ticker, nom)
    if data is not None:
        premier = nom
        break

if data is None:
    print("\n❌ Impossible de télécharger!")
    exit()

print(f"\n✅ Utilisation de: {premier}")

# ===== CALCULER LES INDICATEURS =====

print("\n" + "─"*80)
print("2️⃣  CALCULER LES INDICATEURS")
print("─"*80)

print(f"\n  Moyennes mobiles...", end=" ", flush=True)
data['MA20'] = data['Close'].rolling(window=20).mean()
data['MA50'] = data['Close'].rolling(window=50).mean()
print("✅")

# ===== RSI =====

print(f"  RSI (Relative Strength Index)...", end=" ", flush=True)

def calculer_rsi(d, period=14):
    """Calcule le RSI"""
    delta = d['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    
    # Évite la division par zéro
    rs = np.where(loss != 0, gain / loss, 0)
    rsi = 100 - (100 / (1 + rs))
    
    return pd.Series(rsi, index=d.index)

data['RSI'] = calculer_rsi(data)
print("✅")

# ===== GÉNÉRER LES SIGNAUX =====

print("\n" + "─"*80)
print("3️⃣  GÉNÉRER LES SIGNAUX")
print("─"*80)

print(f"\n  Création des signaux...", end=" ", flush=True)

data['Signal'] = 'HOLD'

# BUY: Prix > MA20 ET MA20 > MA50 ET RSI < 70
buy_condition = (
    (data['Close'] > data['MA20']) & 
    (data['MA20'] > data['MA50']) & 
    (data['RSI'] < 70)
)
data.loc[buy_condition, 'Signal'] = 'BUY'

# SELL: Prix < MA20 ET MA20 < MA50 ET RSI > 30
sell_condition = (
    (data['Close'] < data['MA20']) & 
    (data['MA20'] < data['MA50']) & 
    (data['RSI'] > 30)
)
data.loc[sell_condition, 'Signal'] = 'SELL'

print("✅")

# ===== RÉSULTATS =====

print("\n" + "─"*80)
print("4️⃣  RÉSULTATS DES SIGNAUX")
print("─"*80)

buy_count = len(data[data['Signal'] == 'BUY'])
sell_count = len(data[data['Signal'] == 'SELL'])
hold_count = len(data[data['Signal'] == 'HOLD'])

print(f"\n📊 Nombre de signaux:")
print(f"  🟢 BUY:  {buy_count}")
print(f"  🔴 SELL: {sell_count}")
print(f"  ⚪ HOLD: {hold_count}")

# Signal actuel
signal_actuel = data['Signal'].iloc[-1]
prix_actuel = float(data['Close'].iloc[-1])
ma20_actuel = float(data['MA20'].iloc[-1])
ma50_actuel = float(data['MA50'].iloc[-1])
rsi_actuel = float(data['RSI'].iloc[-1])

print(f"\n💰 Signal Actuel:")
print(f"  Prix: {prix_actuel:.2f}")
print(f"  MA20: {ma20_actuel:.2f}")
print(f"  MA50: {ma50_actuel:.2f}")
print(f"  RSI: {rsi_actuel:.2f}")
print(f"  Signal: {signal_actuel}")

if signal_actuel == 'BUY':
    print(f"\n  🟢 RECOMMANDATION: ACHETER! ✅")
    print(f"     Les conditions techniques sont réunies!")
elif signal_actuel == 'SELL':
    print(f"\n  🔴 RECOMMANDATION: VENDRE! ⚠️")
    print(f"     Les conditions deviennent négatives")
else:
    print(f"\n  ⚪ RECOMMANDATION: ATTENDRE")
    print(f"     Pas de signal clair actuellement")

# ===== DERNIERS JOURS =====

print("\n" + "─"*80)
print("5️⃣  DERNIERS 10 JOURS")
print("─"*80)

affichage = data[['Close', 'MA20', 'MA50', 'RSI', 'Signal']].tail(10)
print("\n" + affichage.to_string())

# ===== STATISTIQUES =====

print("\n" + "─"*80)
print("6️⃣  STATISTIQUES")
print("─"*80)

# Rendement du jour
rendement_jour = ((data['Close'].iloc[-1] - data['Close'].iloc[-2]) / data['Close'].iloc[-2] * 100)

# Rendement depuis le début
rendement_total = ((data['Close'].iloc[-1] - data['Close'].iloc[0]) / data['Close'].iloc[0] * 100)

# Volatilité
volatilite = data['Close'].pct_change().std() * 100

print(f"\n📊 Statistiques:")
print(f"  Rendement du jour: {rendement_jour:+.2f}%")
print(f"  Rendement depuis début: {rendement_total:+.2f}%")
print(f"  Volatilité: {volatilite:.2f}%")
print(f"  Prix min (données): {data['Close'].min():.2f}")
print(f"  Prix max (données): {data['Close'].max():.2f}")

# ===== GRAPHIQUE =====

print("\n" + "─"*80)
print("7️⃣  GRAPHIQUE")
print("─"*80)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Graphique 1: Prix + Signaux
ax1.plot(data.index, data['Close'], linewidth=2, label='Prix', color='black')
ax1.plot(data.index, data['MA20'], linewidth=2, label='MA20', color='orange', alpha=0.7)
ax1.plot(data.index, data['MA50'], linewidth=2, label='MA50', color='red', alpha=0.7)

# Affiche les signaux
buy = data[data['Signal'] == 'BUY']
sell = data[data['Signal'] == 'SELL']

ax1.scatter(buy.index, buy['Close'], color='green', marker='^', s=200, label='BUY Signal', zorder=5)
ax1.scatter(sell.index, sell['Close'], color='red', marker='v', s=200, label='SELL Signal', zorder=5)

ax1.set_title(f'{premier} - Signaux de Trading', fontsize=14, fontweight='bold')
ax1.set_ylabel('Prix')
ax1.legend()
ax1.grid(alpha=0.3)

# Graphique 2: RSI
ax2.plot(data.index, data['RSI'], linewidth=2, color='steelblue', label='RSI')
ax2.axhline(y=70, color='red', linestyle='--', linewidth=1, label='Surachat (70)')
ax2.axhline(y=30, color='green', linestyle='--', linewidth=1, label='Survendu (30)')
ax2.fill_between(data.index, 30, 70, alpha=0.1, color='gray')
ax2.fill_between(data.index, 0, 30, alpha=0.05, color='green')
ax2.fill_between(data.index, 70, 100, alpha=0.05, color='red')

ax2.set_title('RSI - Indicateur de Force', fontsize=12, fontweight='bold')
ax2.set_ylabel('RSI')
ax2.set_xlabel('Date')
ax2.set_ylim(0, 100)
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Graphiques affichés!")

# ===== ANALYSE DÉTAILLÉE =====

print("\n" + "─"*80)
print("8️⃣  ANALYSE DÉTAILLÉE")
print("─"*80)

print(f"\n💡 Interprétation du Signal Actuel:")
print(f"\n  Prix actuel: {prix_actuel:.2f}")
print(f"  MA20 (court terme): {ma20_actuel:.2f}")
print(f"  MA50 (moyen terme): {ma50_actuel:.2f}")
print(f"  RSI: {rsi_actuel:.2f}")

print(f"\n  Conditions:")

if prix_actuel > ma20_actuel:
    print(f"    ✅ Prix au-dessus de MA20")
else:
    print(f"    ❌ Prix sous MA20")

if ma20_actuel > ma50_actuel:
    print(f"    ✅ MA20 au-dessus de MA50 (tendance haussière)")
else:
    print(f"    ❌ MA20 sous MA50 (tendance baissière)")

if rsi_actuel < 30:
    print(f"    🟢 RSI en zone de survendu (ACHAT potentiel)")
elif rsi_actuel > 70:
    print(f"    🔴 RSI en zone de surachat (VENTE potentiel)")
else:
    print(f"    ⚪ RSI en zone normale")

print("\n" + "="*80)
print("✅ JOUR 4 TERMINÉ!")
print("="*80)
```

```python
# ========== JOUR 5: ANALYSEUR COMPLET FINAL ==========

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')

print("\n" + "="*100)
print(" "*20 + "📊 ANALYSEUR FINANCIER COMPLET - VERSION FINALE")
print("="*100)

# ===== FONCTION COMPLÈTE D'ANALYSE =====

def analyser_action_complete(ticker, nom):
    """Analyse complète d'une action"""
    
    print(f"\n  📊 {nom:<15} ({ticker:<6})", end=" ", flush=True)
    
    try:
        # Télécharge
        data = yf.download(ticker, start='2024-01-01', end='2026-04-29', progress=False)
        
        # Vérifications
        if len(data) == 0:
            print("❌ (0 données)")
            return None
        
        # Nettoyage MultiIndex
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = data.columns.get_level_values(0)
        
        # Calculs
        data['MA20'] = data['Close'].rolling(window=20).mean()
        data['MA50'] = data['Close'].rolling(window=50).mean()
        
        # RSI
        delta = data['Close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
        rs = np.where(loss != 0, gain / loss, 0)
        data['RSI'] = 100 - (100 / (1 + rs))
        
        # Signal
        data['Signal'] = 'HOLD'
        data.loc[
            (data['Close'] > data['MA20']) & 
            (data['MA20'] > data['MA50']) & 
            (data['RSI'] < 70),
            'Signal'
        ] = 'BUY'
        data.loc[
            (data['Close'] < data['MA20']) & 
            (data['MA20'] < data['MA50']) & 
            (data['RSI'] > 30),
            'Signal'
        ] = 'SELL'
        
        # Extraction des résultats
        prix = float(data['Close'].iloc[-1])
        ma20 = float(data['MA20'].iloc[-1]) if not pd.isna(data['MA20'].iloc[-1]) else None
        ma50 = float(data['MA50'].iloc[-1]) if not pd.isna(data['MA50'].iloc[-1]) else None
        rsi = float(data['RSI'].iloc[-1]) if not pd.isna(data['RSI'].iloc[-1]) else None
        signal = data['Signal'].iloc[-1]
        
        rendement_jour = ((data['Close'].iloc[-1] - data['Close'].iloc[-2]) / data['Close'].iloc[-2] * 100)
        rendement_total = ((data['Close'].iloc[-1] - data['Close'].iloc[0]) / data['Close'].iloc[0] * 100)
        
        print(f"✅")
        
        return {
            'nom': nom,
            'ticker': ticker,
            'data': data,
            'prix': prix,
            'ma20': ma20,
            'ma50': ma50,
            'rsi': rsi,
            'signal': signal,
            'rendement_jour': rendement_jour,
            'rendement_total': rendement_total
        }
        
    except Exception as e:
        print(f"❌")
        return None

# ===== ANALYSER 5 ACTIONS =====

print("\n" + "─"*100)
print("1️⃣  ANALYSE DE 5 ACTIONS")
print("─"*100)

actions = [
    ('AAPL', 'Apple'),
    ('MSFT', 'Microsoft'),
    ('GOOGL', 'Google'),
    ('AMZN', 'Amazon'),
    ('TSLA', 'Tesla'),
]

resultats = {}

print("\n🔍 Analyse en cours...")

for ticker, nom in actions:
    resultat = analyser_action_complete(ticker, nom)
    if resultat:
        resultats[nom] = resultat

if not resultats:
    print("\n❌ Impossible d'analyser les actions!")
    exit()

print(f"\n✅ {len(resultats)} action(s) analysée(s)")

# ===== TABLEAU RÉCAPITULATIF =====

print("\n" + "─"*100)
print("2️⃣  TABLEAU RÉCAPITULATIF")
print("─"*100)

print(f"\n{'Action':<15} {'Prix':<10} {'MA20':<10} {'MA50':<10} {'RSI':<8} {'Signal':<8} {'Rend%':<10} {'Status':<10}")
print("─"*100)

for nom, resultat in resultats.items():
    prix = f"{resultat['prix']:.2f}"
    ma20 = f"{resultat['ma20']:.2f}" if resultat['ma20'] else "N/A"
    ma50 = f"{resultat['ma50']:.2f}" if resultat['ma50'] else "N/A"
    rsi = f"{resultat['rsi']:.1f}" if resultat['rsi'] else "N/A"
    signal = resultat['signal']
    rend = f"{resultat['rendement_jour']:+.2f}%"
    
    if signal == 'BUY':
        status = "🟢 ACHAT"
    elif signal == 'SELL':
        status = "🔴 VENTE"
    else:
        status = "⚪ ATTENDRE"
    
    print(f"{nom:<15} {prix:<10} {ma20:<10} {ma50:<10} {rsi:<8} {signal:<8} {rend:<10} {status:<10}")

# ===== SIGNAUX IMPORTANTS =====

print("\n" + "─"*100)
print("3️⃣  SIGNAUX IMPORTANTS")
print("─"*100)

buy_signals = {nom: res for nom, res in resultats.items() if res['signal'] == 'BUY'}
sell_signals = {nom: res for nom, res in resultats.items() if res['signal'] == 'SELL'}

if buy_signals:
    print(f"\n🟢 SIGNAUX D'ACHAT ({len(buy_signals)}):")
    for nom, res in buy_signals.items():
        print(f"   • {nom:<15} RSI={res['rsi']:>6.1f}  Rend={res['rendement_jour']:>+6.2f}%")
else:
    print(f"\n⚪ Aucun signal d'achat")

if sell_signals:
    print(f"\n🔴 SIGNAUX DE VENTE ({len(sell_signals)}):")
    for nom, res in sell_signals.items():
        print(f"   • {nom:<15} RSI={res['rsi']:>6.1f}  Rend={res['rendement_jour']:>+6.2f}%")
else:
    print(f"\n⚪ Aucun signal de vente")

# ===== DASHBOARD VISUEL =====

print("\n" + "─"*100)
print("4️⃣  DASHBOARD VISUEL")
print("─"*100)

fig = plt.figure(figsize=(18, 12))
fig.suptitle('📊 ANALYSEUR FINANCIER - DASHBOARD COMPLET', fontsize=16, fontweight='bold')

for idx, (nom, resultat) in enumerate(list(resultats.items())[:6], 1):
    if idx > 6:
        break
    
    ax = plt.subplot(2, 3, idx)
    data = resultat['data']
    
    # Prix + Moyennes
    ax.plot(data.index, data['Close'], linewidth=2, label='Prix', color='black')
    ax.plot(data.index, data['MA20'], linewidth=1.5, label='MA20', color='orange', alpha=0.7)
    ax.plot(data.index, data['MA50'], linewidth=1.5, label='MA50', color='red', alpha=0.7)
    
    # Signaux
    buy = data[data['Signal'] == 'BUY']
    sell = data[data['Signal'] == 'SELL']
    ax.scatter(buy.index, buy['Close'], color='green', marker='^', s=100, zorder=5)
    ax.scatter(sell.index, sell['Close'], color='red', marker='v', s=100, zorder=5)
    
    # Couleur de fond selon signal
    if resultat['signal'] == 'BUY':
        ax.set_facecolor('#e8f5e9')
        titre = f"{nom} 🟢 BUY"
    elif resultat['signal'] == 'SELL':
        ax.set_facecolor('#ffebee')
        titre = f"{nom} 🔴 SELL"
    else:
        titre = f"{nom} ⚪ HOLD"
    
    ax.set_title(titre, fontweight='bold', fontsize=11)
    ax.set_ylabel('Prix', fontsize=9)
    ax.legend(loc='upper left', fontsize=8)
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Graphiques affichés!")

# ===== RAPPORT D'ANALYSE =====

print("\n" + "─"*100)
print("5️⃣  RAPPORT D'ANALYSE DÉTAILLÉ")
print("─"*100)

rapport = []
for nom, res in resultats.items():
    rapport.append({
        'Action': nom,
        'Prix': f"{res['prix']:.2f}",
        'MA20': f"{res['ma20']:.2f}" if res['ma20'] else "N/A",
        'MA50': f"{res['ma50']:.2f}" if res['ma50'] else "N/A",
        'RSI': f"{res['rsi']:.2f}" if res['rsi'] else "N/A",
        'Signal': res['signal'],
        'Rend.Jour': f"{res['rendement_jour']:+.2f}%",
        'Rend.Année': f"{res['rendement_total']:+.2f}%"
    })

rapport_df = pd.DataFrame(rapport)

print("\n" + rapport_df.to_string(index=False))

# ===== RECOMMANDATIONS FINALES =====

print("\n" + "─"*100)
print("6️⃣  RECOMMANDATIONS FINALES")
print("─"*100)

print("""
💡 STRATÉGIE D'INVESTISSEMENT:

1️⃣  COURT TERME (Signaux BUY):
    • Entrez sur les BUY signals
    • Fixez un stop-loss à -2%
    • Prenez profit quand RSI > 70

2️⃣  MOYEN TERME (Tendance):
    • Suivez les croisements MA20/MA50
    • Golden Cross (MA20 > MA50) = HAUSSIER
    • Death Cross (MA20 < MA50) = BAISSIER

3️⃣  GESTION DU RISQUE:
    • Diversifiez entre les actions
    • N'investissez que ce que vous pouvez perdre
    • Utilisez toujours des stop-loss

4️⃣  PROCHAINES ÉTAPES:
    ✅ Semaine 6: Machine Learning & Prédictions
    ✅ Semaine 7: Création d'une App Web
    ✅ Semaine 8: Déploiement en Production
""")

# ===== RÉSUMÉ FINAL =====

print("\n" + "="*100)
print("✅ SEMAINE 5 TERMINÉE!")
print("="*100)

print(f"""
📊 RÉSUMÉ COMPLET:

Actions analysées: {len(resultats)}
Signaux BUY: {len(buy_signals)}
Signaux SELL: {len(sell_signals)}

🎯 Prochaines étapes:
  ✅ Semaine 5: ✓ COMPLÉTÉE
  → Semaine 6: Machine Learning & Prédictions IA
  → Semaine 7: App Web Bloomberg-like
  → Semaine 8: Déploiement Production

💪 Tu es maintenant un Data Analyst Financier!

Analyse effectuée: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
""")

print("="*100 + "\n")
```

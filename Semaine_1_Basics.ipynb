```python
# ========== JOUR 2: OPÉRATIONS MATHÉMATIQUES ==========

# Données de base
prix_achat = 1500      # Prix d'achat d'une action
quantite = 10          # Nombre d'actions achetées
prix_actuel = 1650     # Prix aujourd'hui

# OPÉRATION 1: Multiplication (*)
# Calcule le total investi
total_investi = prix_achat * quantite
print("Total investi:", total_investi, "FCFA")

# OPÉRATION 2: Soustraction (-)
# Calcule le gain par action
gain_par_action = prix_actuel - prix_achat
print("Gain par action:", gain_par_action, "FCFA")

# OPÉRATION 3: Multiplication pour le gain total
gain_total = gain_par_action * quantite
print("Gain total:", gain_total, "FCFA")

# OPÉRATION 4: Division (/)
# Calcule le rendement en %
rendement_pourcent = (gain_par_action / prix_achat) * 100
print("Rendement %:", rendement_pourcent, "%")
```

```python
# ===== SIMULE TON ACHAT =====

# Tu achètes une action à ce prix
mon_prix_achat = 1500

# Tu achètes cette quantité
ma_quantite = 10

# Aujourd'hui, le prix est monté à
mon_prix_actuel = 1650

# À TOI DE CALCULER (sans utiliser ma_quantite * mon_prix_achat directement):
# 1. Total investi
Total_investi = mon_prix_achat * ma_quantite 

# 2. Gain par action
Gain_par_action = mon_prix_actuel - mon_prix_achat
# 3. Gain total
Gain_total = (mon_prix_actuel*ma_quantite) - Total_investi
# 4. Rendement %

Rendement = (Gain_par_action/mon_prix_achat)*100

# Écris tes calculs ici:
print("Mon total investi est :", Total_investi)
print("Mon gain par action est :", Gain_par_action)
print("Mon gain total est :", Gain_total)
print("Mon rendement en % est :", Rendement)
```

```python
# ========== JOUR 3: CONDITIONS ==========

# Scénario: Dois-je vendre mon action Orange?

prix_achat = 1500
prix_actuel = 1600
mon_objectif_vente = 1650

# CONDITION 1: Le prix a-t-il atteint mon objectif?
if prix_actuel >= mon_objectif_vente:
    print("✅ VENDRE! Objectif atteint!")
else:
    print("⏳ ATTENDRE. Objectif non atteint.")

print("\n" + "="*50 + "\n")

# CONDITION 2: Suis-je en gain ou en perte?
if prix_actuel > prix_achat:
    print("📈 EN GAIN! Le prix a monté!")
else:
    print("📉 EN PERTE. Le prix a baissé.")
```

```python
# Scénario plus réaliste: Quelle action prendre?

prix_actuel = 1600
prix_achat = 1500
gain = prix_actuel - prix_achat

if gain > 200:
    print("🚀 GAIN ÉNORME! VENDRE MAINTENANT!")
elif gain > 100:
    print("📈 BON GAIN. Tu peux vendre ou attendre.")
elif gain > 0:
    print("📊 PETIT GAIN. À toi de décider.")
else:
    print("❌ EN PERTE. NE PAS VENDRE! Attends la remontée.")
```

```python
# Condition avec AND (ET): Les 2 doivent être VRAIS

prix_actuel = 1650
volume_echange_bon = True
tendance_haussiere = True

if prix_actuel > 1600 and volume_echange_bon and tendance_haussiere:
    print("✅ TOUS LES SIGNAUX SONT VERTS! C'est le moment!")
else:
    print("⚠️ Un signal n'est pas bon.")

print("\n" + "="*50 + "\n")

# Condition avec OR (OU): AU MOINS UN doit être VRAI

prix_tres_bon = False
action_en_promotion = True

if prix_tres_bon or action_en_promotion:
    print("💡 Au moins un signal positif. Considère l'achat.")
else:
    print("❌ Aucun signal positif.")
```

```python
# ===== TON SYSTÈME DE DÉCISION =====

# Tes seuils personnels
mon_prix_achat = 1500
mon_prix_vente_ideale = 1650
mon_prix_perte_limite = 1450

# Le prix actuel (change-le pour tester)
prix_du_jour = 1600

gain = prix_du_jour - mon_prix_achat

# À TOI D'ÉCRIRE LA LOGIQUE:
# SI prix_du_jour >= mon_prix_vente_ideale:  VENDRE!

if prix_du_jour >= mon_prix_vente_ideale:
    print("✅ VENDRE! Objectif atteint!")
elif prix_du_jour > mon_prix_achat:
    print("📈 BON GAIN; ATTENDRE")
elif prix_du_jour > mon_prix_perte_limite:
    print("EN PERTE LÉGÈRE, ATTENDRE")
else:
    print("PERTE IMPORTANTE, STOP-LOSS (vendre pour limiter la casse).")
```

```python
# ========== JOUR 4: MISE EN PRATIQUE ==========
# Combine VARIABLES + OPÉRATIONS + CONDITIONS

print("="*60)
print("MON ANALYSEUR D'ACTIONS PERSONNEL")
print("="*60)

# ===== MES DONNÉES =====
nom_action = "SONATEL"
prix_achat = 29106
quantite = 10
prix_actuel = 28400

# ===== CALCULS =====
total_investi = prix_achat * quantite
total_actuel = prix_actuel * quantite
gain_total = total_actuel - total_investi
rendement_pourcent = (gain_total / total_investi) * 100

# ===== AFFICHAGE =====
print(f"\n📊 FICHE ACTION: {nom_action}")
print(f"─" * 60)
print(f"Prix d'achat:        {prix_achat} FCFA")
print(f"Quantité:            {quantite}")
print(f"Total investi:       {total_investi} FCFA")
print(f"\nPrix actuel:         {prix_actuel} FCFA")
print(f"Total actuel:        {total_actuel} FCFA")
print(f"Gain/Perte:          {gain_total} FCFA")
print(f"Rendement:           {rendement_pourcent:.2f}%")
print(f"─" * 60)

# ===== DÉCISION =====
if rendement_pourcent >= 10:
    action = "✅ VENDRE! (Rendement excellent)"
elif rendement_pourcent >= 5:
    action = "📈 BON! (Rendement satisfaisant)"
elif rendement_pourcent > 0:
    action = "📊 PETIT GAIN. (Attendre peut-être)"
else:
    action = "❌ EN PERTE. (Attendre la remontée)"

print(f"CONSEIL:             {action}")
print("="*60)
```

```python
# ========== JOUR 5: MON PORTEFEUILLE BRVM ==========

print("="*70)
print("PORTEFEUILLE D'ACTIONS - BOURSE RÉGIONALE DES VALEURS MOBILIÈRES")
print("="*70)

# ===== TES ACTIONS (Change avec tes vraies données!) =====

# Action 1
action1_nom = "Orange"
action1_prix_achat = 1500
action1_quantite = 10
action1_prix_actuel = 1650

# Action 2
action2_nom = "Safarel"
action2_prix_achat = 2000
action2_quantite = 5
action2_prix_actuel = 1950

# Action 3
action3_nom = "Ecobank"
action3_prix_achat = 5000
action3_quantite = 2
action3_prix_actuel = 4800

# ===== CALCULS POUR CHAQUE ACTION =====

print(f"\n{'Entreprise':<15} {'Prix':<10} {'Quantité':<10} {'Total':<15} {'Gain/Perte':<15} {'%':<8}")
print("─"*70)

# Action 1
total1 = action1_prix_achat * action1_quantite
total_actuel1 = action1_prix_actuel * action1_quantite
gain1 = total_actuel1 - total1
rendement1 = (gain1 / total1) * 100

print(f"{action1_nom:<15} {action1_prix_achat:<10} {action1_quantite:<10} {total1:<15} {gain1:<15.0f} {rendement1:<8.2f}%")

# Action 2
total2 = action2_prix_achat * action2_quantite
total_actuel2 = action2_prix_actuel * action2_quantite
gain2 = total_actuel2 - total2
rendement2 = (gain2 / total2) * 100

print(f"{action2_nom:<15} {action2_prix_achat:<10} {action2_quantite:<10} {total2:<15} {gain2:<15.0f} {rendement2:<8.2f}%")

# Action 3
total3 = action3_prix_achat * action3_quantite
total_actuel3 = action3_prix_actuel * action3_quantite
gain3 = total_actuel3 - total3
rendement3 = (gain3 / total3) * 100

print(f"{action3_nom:<15} {action3_prix_achat:<10} {action3_quantite:<10} {total3:<15} {gain3:<15.0f} {rendement3:<8.2f}%")

# ===== TOTAUX =====
print("─"*70)
total_investi = total1 + total2 + total3
total_actuel = total_actuel1 + total_actuel2 + total_actuel3
gain_total = total_actuel - total_investi
rendement_total = (gain_total / total_investi) * 100

print(f"{'TOTAL':<15} {'':<10} {'':<10} {total_investi:<15.0f} {gain_total:<15.0f} {rendement_total:<8.2f}%")
print("="*70)

# ===== RÉSUMÉ =====
print(f"\n💰 Total investi:     {total_investi} FCFA")
print(f"💵 Valeur actuelle:   {total_actuel} FCFA")
print(f"📈 Gain/Perte total:  {gain_total} FCFA")
print(f"📊 Rendement global:  {rendement_total:.2f}%")

if gain_total > 0:
    print("\n✅ TON PORTEFEUILLE EST EN GAIN!")
else:
    print("\n⚠️ TON PORTEFEUILLE EST EN PERTE. Attends la remontée.")

print("="*70)
```
